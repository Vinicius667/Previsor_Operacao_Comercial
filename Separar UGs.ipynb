```python
import pandas as pd
import re
```

```python
bdworgs_df = pd.read_excel("./bdworgs.xlsx")
bdworgs_df = bdworgs_df.astype(dtype=str)
```

## O código abaixo comprova que não há números separados por espaços

```python
ugs = bdworgs_df["Num_UG"].to_list()
for ug in ugs:
    if len(re.findall(r"\d\s+\d",ug)) > 0:
        print(ug)
```

## O código abaixo demonstra a conversão entre formatos da coluna Num_UG

```python
def corrige_ug(ug):
    ug = ug.replace(" ","")
    ug = ug.replace("e",",")
    ug = ug.replace("-","a")
    ug = re.sub(r",+",r",",ug)
    ug = re.sub(r"([,ae])(0)",r"\1",ug)
    ug = re.sub(r"^0",r"",ug)
    ug = ug.split("a")
    lista = []
    for i in range(len(ug)-1):
        menor = int(ug[i].split(",")[-1])
        maior = int(ug[i+1].split(",")[0])
        list_nums = list(range(menor+1,maior))
        lista += list_nums
    ug = [el.split(",") for el in ug]
    ug = [int(el2) for el1 in ug for el2 in el1]
    ug+= lista
    return (sorted(ug))

with open("bdworgs_ugs.txt","w") as f:
    for i,ug in enumerate(ugs):
        ug_corr = corrige_ug(ug)
        f.write(f"{ug} => {ug_corr}\n")

corrige_ug("01 a 04 e 06 a 08, 11,15 a 18")
```

## O código abaixo demonstra a conversão entre formatos da coluna Num_UG

```python
bdworgs_df.insert(3,"Quant_UG",bdworgs_df["Num_UG"].apply(lambda x : len(corrige_ug(x)))) 
```
